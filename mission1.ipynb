````python
# 2
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df_2011 = df[df['Year'] >= 2011] # Year가 2011이상인 행만 가져오기
result = df_2011.groupby(['Year', 'Country'])['Life_Expectancy'].mean().unstack()
# groupby로 년, 국가로 그룹화 -> 수명 평균내기 -> 컬럼에 인덱스값을 올림(unstack)
result
````
