```python
df_raw = pd.read_csv("국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20221201103047.csv", encoding="cp949")
```

```python
#melt함수 
```

```python
df = df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'], var_name='기간',value_name='100만원')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
#데이터전처리 
#기간에서 연도를 분리 
```

```python
df['기간'].unique()
```

```python
df['연도'] = df['기간'].map(lambda x : int(x.split('.')[0]))
```

```python
df.head()
```

```python
df['분기'] = df['기간'].map(lambda x : int(x.split('.')[1].split('/')[0]))
df.head()
```

```python
#데이터타입 변경하기
# - 문자를 결측치로(NaN) 변경하고 float 타입으로 변경 
df['100만원'] = df['100만원'].replace('-', np.nan).astype(float)
df.head()
```

```python
# 합계데이터는 제거 한다 
df = df[(df['국가(대륙)별']!='합계')&(df['상품군별']!='합계')].copy()
```

```python
df.isnull().sum()
```

```python
#소계를 가져와서..
df_total = df[df['판매유형별']=='계'].copy()
df_total.head()
```
